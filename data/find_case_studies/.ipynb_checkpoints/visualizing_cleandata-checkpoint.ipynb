```python
# Create dataframe
df = pd.read_sql_query('select * from ' + tablename, con=conn)
```

```python
# Narrow it down to the variables we want to take a look at
X = pd.DataFrame(df[['landbaseyear', 'yearbuilt', 'effectiveyearbuilt','landvalue','sqftmain']])

# Preliminary cleaning
for col in X:
    # Removes any row where column value is ''
    X = X[X[col]!= ''] 
    
    # Change data type to int and float
    if col == 'landbaseyear':
        X[col] = X[col].astype(int)
    else:
        X[col] = X[col].astype(float)

# Add land value per square foot to dataframe
X['landvaluepersqft'] = X['landvalue']/X['sqftmain']
```

```python
# Plot distribution of land values
plt.hist(X['landvalue'], bins=100, range=(0,10000000));
plt.title('Distribution of Land Values (cut off at $10m)');
```
